```python
#Aramamos listado de fechas
fechas = []
for i in range((datetime.today().replace(year=datetime.today().date().year + 1).date() -
                datetime.today().date()).days):
    fechas.append("'" + (datetime.today().date()
                         + timedelta(days=i)).strftime('%Y-%m-%d') + "'")
fechas
```

```python
evento = pd.read_excel('explotar jugueteria impo.xlsx')
evento.columns = evento.columns.str.upper()
evento.columns = evento.columns.str.strip()
evento
```

```python
cols = {'ITEM':'ORIN', 'LOCAL':'GEOG_LOCL_COD', 'LOCALES':'GEOG_LOCL_COD', 'EVENTO':'EVENTO_ID'}
```

```python
evento.rename(columns=cols, inplace=True)
```

```python
evento.info()
```

```python
if 'GEOG_LOCL_COD' in evento.columns:
    if 'EVENTO_ID' in evento.columns:
        evento = evento[['ORIN', 'GEOG_LOCL_COD', 'EVENTO_ID']].drop_duplicates().astype('str')
    else:
        evento = evento[['ORIN', 'GEOG_LOCL_COD']].drop_duplicates().astype('str')
else:
    evento = evento[['ORIN']].drop_duplicates()
    evento['aux'] = 1
    locales = pd.read_excel('LOCALES.xlsx')
    locales['aux'] = 1
    evento = evento.merge(locales).drop(columns='aux').astype('str')
evento
```

```python
evento['GEOG_LOCL_COD'] = evento['GEOG_LOCL_COD'].str.replace(';', ',')
evento
```

```python
evento['GEOG_LOCL_COD'] = evento['GEOG_LOCL_COD'].str.split(',')
evento
```

```python
evento = evento.explode(column='GEOG_LOCL_COD')
evento
```

```python
evento.info()
```

```python
evento['GEOG_LOCL_COD'] = evento['GEOG_LOCL_COD'].str.strip()
evento['ORIN'] = evento['ORIN'].str.strip()
evento
```

```python
def limpLoc (l:str):
    if len(l) == 3:
        return l
    else:
        return pd.NA
```

```python
evento['GEOG_LOCL_COD'] = evento['GEOG_LOCL_COD'].apply(limpLoc)
evento[evento['GEOG_LOCL_COD'].isna()]
```

```python
evento.dropna(inplace=True)
evento[evento['GEOG_LOCL_COD'].isna()]
```
